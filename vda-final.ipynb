```python
# Định nghĩa hàm xác định trạng thái (tăng, giảm, không đổi)
def get_trend(val1, val2):
    if val1 > val2:
        return "giảm"
    elif val1 < val2:
        return "tăng"
    else:
        return "không đổi"
```

```python
def plot_heatmap(df, columns, name):
    # Tính ma trận tương quan
    corr_matrix = df[columns].corr()
    
    # Vẽ heatmap
    plt.figure(figsize=(8, 6))
    sns.heatmap(corr_matrix, annot=True, cmap='OrRd', vmin=0, vmax=1)
    plt.title(f'Correlation for {name}')
    plt.xticks(rotation=90, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

def plot_heatmap_total(df, columns):
    # Tính ma trận tương quan
    corr_matrix = df[columns].corr()
    
    # Vẽ heatmap
    plt.figure(figsize=(24, 18))
    sns.heatmap(corr_matrix, annot=False, cmap='OrRd', vmin=0, vmax=1)
    plt.title('Correlation for Total Group')
    plt.xticks(rotation=90, ha='right')
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

loan_features    = [c for c in df_winsorized.columns if 'NUM_NEW_LOAN_TAKEN' in c]
balance_features = [c for c in df_winsorized.columns if 'OUTSTANDING_BAL'    in c]
enquiry_features = [c for c in df_winsorized.columns if 'ENQUIRIES'          in c]

scaler  = StandardScaler()

# --- Loan group: PCA -> 3 components ---
loan_imp   = df_winsorized[loan_features]
loan_scaled= scaler.fit_transform(loan_imp)
pca_loan   = PCA(n_components=3, random_state=42)
loan_pca   = pca_loan.fit_transform(loan_scaled)
loan_pca_df= pd.DataFrame(
    loan_pca,
    columns=[f'NUM_NEW_LOAN_TAKEN_PCA_{i+1}' for i in range(3)],
    index=df_winsorized.index
)

# --- Balance group: PCA -> 5 components ---
bal_imp      = df_winsorized[balance_features]
bal_scaled   = scaler.fit_transform(bal_imp)
pca_balance  = PCA(n_components=5, random_state=42)
balance_pca  = pca_balance.fit_transform(bal_scaled)
balance_pca_df = pd.DataFrame(
    balance_pca,
    columns=[f'OUTSTANDING_BAL_PCA_{i+1}' for i in range(5)],
    index=df_winsorized.index
)

# --- Enquiry group: PCA -> 5 components ---
enq_imp      = df_winsorized[enquiry_features]
enq_scaled   = scaler.fit_transform(enq_imp)
pca_enquiry  = PCA(n_components=5, random_state=42)
enquiry_pca  = pca_enquiry.fit_transform(enq_scaled)
enquiry_pca_df = pd.DataFrame(
    enquiry_pca,
    columns=[f'ENQUIRIES_PCA_{i+1}' for i in range(5)],
    index=df_winsorized.index
)

df_pca = pd.concat(
    [
        df_winsorized[['label', "SHORT_TERM_COUNT", "MID_TERM_COUNT", "SHORT_TERM_COUNT_BANK", "MID_TERM_COUNT_BANK", "SHORT_TERM_COUNT_NON_BANK", "MID_TERM_COUNT_NON_BANK", "NUMBER_OF_LOANS", "NUMBER_OF_LOANS_BANK", "NUMBER_OF_LOANS_NON_BANK", "NUMBER_OF_CREDIT_CARDS", "NUMBER_OF_CREDIT_CARDS_BANK", "NUMBER_OF_CREDIT_CARDS_NON_BANK", "NUMBER_OF_RELATIONSHIP", "NUMBER_OF_RELATIONSHIP_BANK", "NUMBER_OF_RELATIONSHIP_NON_BANK", "INCREASING_BAL_3M_LOAN", "INCREASING_BAL_6M_LOAN", "INCREASING_BAL_3M_CC", "INCREASING_BAL_6M_CC", "INCREASING_BAL_3M_ALL", "INCREASING_BAL_6M_ALL", "CREDIT_CARD_NUMBER_OF_LATE_PAYMENT"]],
        loan_pca_df,
        balance_pca_df,
        enquiry_pca_df
    ], axis=1
)

print(df_pca.head())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_X, y, test_size=0.2, random_state=42)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
from sklearn.linear_model import LogisticRegression

model1 = LogisticRegression(penalty='l2', tol=1e-6, random_state=42, solver='liblinear', max_iter=300)

model1.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier()
params2 = {
    'n_estimators': [100, 150, 200],
    'criterion': ['gini', 'entropy'],
    'max_depth': [20, 35, 50],
    'max_features': ['sqrt', 'log2']
}

search2 = RandomizedSearchCV(
    estimator = model2,
    param_distributions=params2,
    n_iter = 20,
    scoring='f1_weighted',
    n_jobs=-1,
    cv=3
)
search2.fit(X_train, y_train)
print(search2.best_params_)
print(search2.best_score_)
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import RandomizedSearchCV
model3 = XGBClassifier()

params3 = {
    'n_estimators': [100, 150, 200],
    'learning_rate': [0.01, 0.05, 0.1],
    'max_depth': [20, 35, 50],
    'reg_alpha': [1, 5, 10],
    'reg_lambda': [0.1, 0.5, 1]
}
search3 = RandomizedSearchCV(
    estimator=model3,
    param_distributions=params3,
    n_iter=20,         
    scoring='f1_weighted',
    cv=3,              
    verbose=1,
    random_state=42,
    n_jobs=-1
)
search3.fit(X_train, y_train)
print(search3.best_params_)
print(search3.best_score_)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
def prediction(model):
    y_preds = model.predict(X_test)
    print(f"Model: {model}")
    print(f"Accuracy: \n{accuracy_score(y_preds, y_test)}")
    print(f"Confusion Matrix: \n{confusion_matrix(y_test, y_preds)}")
    print(f"Classification Report: \n{classification_report(y_preds, y_test)}")
```

```python
prediction(model1)
prediction(search2)
prediction(search3)
```

```python
from sklearn.metrics import roc_auc_score

def gini(model):
    if hasattr(model, 'predict_proba'):
        prediction = model.predict_proba(X_test)[:, 1]
    else:
        prediction = model.predict(X_test)
    auc_model = roc_auc_score(y_test, prediction)
    return auc_model * 2 - 1

gini_1 = gini(model1)
gini_2 = gini(search2)
gini_3 = gini(search3)

print(f"Gini Index of Logistic Regression: {gini_1}")
print(f"Gini Index of Random Forest: {gini_2}")
print(f"Gini Index of XGBoost: {gini_3}")
```

```python
from sklearn.metrics import recall_score
def recall(models):
    recalls = []
    for model in models:
        prediction = model.predict(X_test)
        get_recall = recall_score(y_test, prediction)
        recalls.append(get_recall)
    return recalls

def accuracy(models):
    accuracies = []
    for model in models:
        prediction = model.predict(X_test)
        get_acu = accuracy_score(y_test, prediction)
        accuracies.append(get_acu)
    return accuracies

models = [model1, search2, search3]
models_name = ['Logistic Regression', 'Random Forest', 'XGBoost']
ginis = [gini_1, gini_2, gini_3]
recalls = recall(models)
accuracies = accuracy(models)
```

```python
x = np.arange(len(models_name))
width = 0.2

fig, ax = plt.subplots(figsize=(10, 6))

bars1 = ax.bar(x - width, ginis, width=width, color='orange', label='Gini')
bars2 = ax.bar(x, recalls, width=width, color='blue', label='Recall')
bars3 = ax.bar(x + width, accuracies, width=width, color='red', label='Accuracy')

for bars in [bars1, bars2, bars3]:
    ax.bar_label(bars, fmt='%.2f', padding=3)

ax.set_xticks(x)
ax.set_xticklabels(models_name)
ax.set_xlabel('Models')
ax.set_ylabel('Scores') 
ax.set_title('Model Performance Comparison')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
from optbinning import OptimalBinning
bin_intervals = []

for col in X.columns:
    optb = OptimalBinning(name=col, dtype="numerical", monotonic_trend="auto_heuristic")
    optb.fit(X[col], y)
    table = optb.binning_table.build()  
    
    for _, row in table.iterrows():
        bin_intervals.append({
            "Feature": col,
            "Information Values": optb.binning_table.iv,
            "Bin Range": row["Bin"],  
            "WoE": row["WoE"]         
        })

bin_intervals = pd.DataFrame(bin_intervals)

df_iv = pd.DataFrame(bin_intervals).sort_values('Information Values', ascending=False)
df_iv = df_iv.loc[(df_iv['Information Values'] >= 0.1) & (df_iv['Information Values'] < 1)]
df_iv
```

```python
outlier = df_iv.loc[(df_iv['Bin Range'] == "") | (df_iv['Bin Range'] == "Special") | (df_iv['Bin Range'] == "Missing")]
df_iv = df_iv.drop(outlier.index)
df_iv
```

```python
columns = np.unique(df_iv['Feature'])   
cus_id = df['customer_id']  
label = df['label']


feature_data = X[list(columns)]

customers = pd.concat([cus_id.reset_index(drop=True), feature_data.reset_index(drop=True), label.reset_index(drop=True)], axis=1)
customers.columns = ['customer_id'] + list(columns) + ['label']
customers
```

```python
def getInterval(string):
    interval_str = string.replace("(", "").replace(")", "").replace("[", "")
    left, right = interval_str.split(", ")
    return left, right

df_iv[["Bin Left", "Bin Right"]] = df_iv['Bin Range'].apply(lambda x: pd.Series(getInterval(str(x))))
df_iv.head()
```

```python
df_iv['Bin Left'] = df_iv['Bin Left'].str.replace("inf", "100000000")
df_iv['Bin Left'] = df_iv['Bin Left'].str.replace("-inf", "-100000000")
df_iv['Bin Right'] = df_iv['Bin Right'].str.replace("inf", "100000000")
df_iv
```

```python
df_iv['Bin Left'] = pd.to_numeric(df_iv['Bin Left'], errors='coerce')
df_iv['Bin Right'] = pd.to_numeric(df_iv['Bin Right'], errors='coerce')
```

```python
def get_woe(value, bins):
    for _, row in bins.iterrows():
        if row['Bin Left'] <= value < row['Bin Right']:
            return row['WoE']
    return np.nan  

for feature in df_iv['Feature'].unique():
    bins = df_iv[df_iv['Feature'] == feature][['Bin Left', 'Bin Right', 'WoE']]
    customers[feature + '_WoE'] = customers[feature].apply(lambda x: get_woe(x, bins))
customers
```

```python
columns_to_move = ['label']
customers = customers[[col for col in customers.columns if col not in columns_to_move] + columns_to_move]
customers
```

```python
X_woe = customers.iloc[:, 24:-1]
y_woe = customers['label']
# X_woe
y_woe
```

```python
np.count_nonzero(y_woe == 0)
```

```python
X_woe_arr = np.array(X_woe)
y_woe_arr = np.array(y_woe)
```

```python
LR = LogisticRegression(penalty='l2', tol=1e-5, random_state=42, max_iter=100)

LR.fit(X_woe_arr, y_woe_arr)
```

```python
import numpy as np

base_score = 650
PDO = 50
intercept = LR.intercept_ 
slope = LR.coef_ 
if slope.ndim == 1:
    slope = slope.reshape(-1, 1)
elif slope.shape[0] == 1:
     slope = slope.T 


B = PDO / np.log(2)


base_score_formula_term = base_score - B * intercept

sum_coef_woe_per_sample = - X_woe @ slope

sum_of_feature = B * sum_coef_woe_per_sample

total_scores = base_score_formula_term + sum_of_feature
total_scores
```

```python
customers['Scoreboard'] = total_scores
customers
```
